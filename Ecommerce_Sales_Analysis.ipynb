```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

```python
#Task 1: Basic Grouping and Single Aggregations

#Perform the following grouping operations:

    #Calculate total sales amount for each region using groupby() and sum()
    #Group by region and calculate sum of sales_amount

region_sales = df.groupby('region')['sales_amount'].sum().reset_index()

print("\nTotal Sales by Region:")
print(region_sales)
```

df.groupby('region'): This groups the DataFrame df by the unique values in the region column. This means all rows with the same region will be grouped together.
['sales_amount'].sum(): After grouping, it selects the sales_amount column and then calculates the sum of sales_amount for each group (i.e., for each region).

```python
#Sort Regions by Highest Sales
region_sales_sorted = region_sales.sort_values(by='sales_amount', ascending=False)

print("\nTop Performing Regions:")
print(region_sales_sorted)
```



Certainly! This code sorts the region_sales DataFrame, which we previously created, to identify the top-performing regions. Here's how it works:

    region_sales.sort_values(by='sales_amount', ascending=False): This line sorts the region_sales DataFrame.
        by='sales_amount': Specifies that the sorting should be based on the values in the sales_amount column.
        ascending=False: Indicates that the sorting should be in descending order, meaning the regions with the highest sales will appear at the top.


```python
#Count Number of Transactions for Each Category
category_transactions = df.groupby('category')['transaction_id'].count().reset_index()

print("\nTransactions per Category:")
print(category_transactions)
```

df.groupby('category'): This groups the DataFrame df based on the unique values in the category column. All rows belonging to the same category are grouped together.
['transaction_id'].count(): After grouping, it selects the transaction_id column. The .count() method then tallies the number of non-null transaction IDs within each category group

```python
#Task 2: Multi-Column Grouping and Multiple Aggregations
#Total Sales for Each Region and Category Combination

#Group by region and category together and calculate the total sales amount.

region_category_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()

print("\nTotal Sales by Region and Category:")
print(region_category_sales)
```


    df.groupby(['region', 'category']): This is a multi-column grouping. It groups the DataFrame df based on the unique pairs of values found in both the region and category columns. All rows that share the same region and category will be grouped together.
    ['sales_amount'].sum(): After the data has been grouped by both region and category, this part selects the sales_amount column and then calculates the sum of sales_amount for each of those unique region-category groups. This gives you the total sales for each specific region-category combination.
    .reset_index(): Similar to previous examples, after groupby() and sum(), region and category would become the multi-level index of the resulting Series. reset_index() converts these indices back into regular columns in the DataFrame, making it easier to read and use.

The final two lines print a descriptive header and then display the region_category_sales DataFrame, showing the total sales for each region and category combination.

```python
#Multiple Metrics for Each Salesperson
#Calculate three metrics at once using .agg().
salesperson_performance = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()

print("\nSalesperson Performance Metrics:")
print(salesperson_performance)
```

df.groupby('salesperson'): This groups the DataFrame df by the unique values in the salesperson column. All rows belonging to the same salesperson are grouped together.
['sales_amount']: After grouping, we select the sales_amount column because that's the column we want to perform aggregations on.
.agg(['sum', 'mean', 'count']): This is where multiple aggregations happen. The .agg() method allows you to apply several aggregation

```python
#Identify the Top Salesperson

#Sort by total sales in descending order.
top_salespersons = salesperson_performance.sort_values(by='sum', ascending=False)

print("\nSalespersons Ranked by Total Sales:")
print(top_salespersons)
```

The final two lines print a descriptive header and then display the top_salespersons DataFrame, showing the salespersons ranked from highest to lowest total sales.

```python
#Find the Category With Maximum Total Revenue

#First calculate total revenue per category, then use .idxmax().
category_sales = df.groupby('category')['sales_amount'].sum()

top_category = category_sales.idxmax()

print("\nTotal Sales by Category:")
print(category_sales)

print("\nCategory with Highest Revenue:")
print(top_category)
```

```python
#Task 3: Custom Aggregation and Complete Sales Report
#Define a Custom Aggregation Function
def sales_range(x):
    return x.max() - x.min()
    print('custom aggregation created:')


```

Create a function to calculate the sales range (difference between maximum and minimum sales).
x represents the grouped column values.

x.max() → highest transaction value

x.min() → lowest transaction value

Range = max − min

```python
#Apply Custom and Standard Aggregations
region_analysis = df.groupby('region')['sales_amount'].agg(
    ['sum', 'mean', 'min', 'max', sales_range]
).reset_index()

print("\nRegional Sales Analysis:")
print(region_analysis)
```

group the dataset by region and apply multiple aggregations including the custom function.
sum → total revenue per region

mean → average transaction value

min / max → lowest and highest sale

sales_range → spread of transaction values

```python
#Create the Final Sales Summary Report

#Now generate a manager-friendly summary report.

final_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

print("\nFinal Regional Sales Summary:")
print(final_report)
```

df.groupby('region'): This groups the DataFrame df by the unique values in the region column. All transactions from the same region are grouped together.
.agg({ 'sales_amount': ['sum', 'mean'], 'customer_id': 'count' }): This is a powerful way to apply different aggregation functions to different columns simultaneously:

    For the sales_amount column, it calculates both the 'sum' (total sales) and the 'mean' (average sales) for each region.
    For the customer_id column, it calculates the 'count'. In this context, using count on customer_id after grouping by region effectively counts the number of transactions (or rows) within each region, assuming customer_id is always present for a transaction.

.reset_index(): After the aggregations, 'region' would become the index. reset_index() converts 'region' back into a regular column, making the output a cleaner DataFrame.
