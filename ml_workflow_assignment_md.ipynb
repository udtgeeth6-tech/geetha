**Data Cleaning & Preprocessing **

This step ensures the data is accurate and model-ready by handling missing values, encoding variables, and scaling features; poor-quality data will lead to unreliable model performance regardless of model complexity.

Why Each Step Matters Before Building a Complex Model
1. **Define the problem clearly**
Before writing a single line of code, you need to confirm what "success" looks like — in this case, a binary classification (repeat purchase: yes or no). Without this, you risk optimizing for the wrong thing entirely.
2. **Inspect and clean the data**
Real-world retail data is messy — missing values, duplicates, outliers. A complex model trained on dirty data will learn noise instead of patterns, and garbage in genuinely means garbage out.
3. **Exploratory Data Analysis (EDA)**
Understanding the distribution of your features and the balance of your target class (how many 1s vs 0s?) prevents surprises later.
4. **Start with a simple baseline model**
A logistic regression or even a majority-class predictor gives you a performance floor.
5.** Feature engineering and selection**
Raw columns aren't always the most informative inputs.
6.** Evaluate properly**
Choosing the right metric (precision, recall, F1, AUC-ROC) before scaling up ensures you're measuring what actually matters to the business, not just what looks good on paper.
