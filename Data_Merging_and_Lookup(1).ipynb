```python
#all in one:final program
import pandas as pd

# ===============================
# === TASK 1: Data Preparation ===
# ===============================

print("=== TASK 1: Data Preparation ===\n")

# Create Students DataFrame
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None,
              'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

students_df = pd.DataFrame(students_data)

print("Original Students DataFrame:")
print(students_df)
print()

# Null Value Analysis
print("Null Value Analysis:")
for col in students_df.columns:
    null_count = students_df[col].isnull().sum()
    null_percent = (null_count / len(students_df)) * 100
    print(f"Column: {col}, Nulls: {null_count} ({null_percent:.2f}%)")

# Cleaning
students_df['city'] = students_df['city'].fillna('Unknown')
students_df = students_df.dropna(subset=['name'])

print("\nCleaned Students DataFrame ({} rows):".format(len(students_df)))
print(students_df)
print()


# ===============================
# === TASK 2: Join Operations ===
# ===============================

print("=== TASK 2: Join Operations ===\n")

# Create Enrollments DataFrame
enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python',
                    'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01',
                        '2024-02-10', '2024-02-15', '2024-03-01']
}

enrollments_df = pd.DataFrame(enrollments_data)

# -------- INNER JOIN --------
inner_df = pd.merge(students_df, enrollments_df,
                    on='student_id', how='inner')

print("Inner Join Result ({} rows):".format(len(inner_df)))
print(inner_df[['student_id', 'name', 'course_name', 'enrollment_date']])

excluded = list(set(students_df['student_id']) -
                set(inner_df['student_id']))

print(f"Excluded students: {excluded} - Not in enrollments table\n")


# -------- LEFT JOIN --------
left_df = pd.merge(students_df, enrollments_df,
                   on='student_id', how='left')

print("Left Join Result ({} rows):".format(len(left_df)))
print(left_df[['student_id', 'name', 'course_name', 'enrollment_date']])

null_courses = left_df[left_df['course_name'].isnull()]['student_id'].tolist()
print(f"Students with null course_name: {null_courses}\n")


# -------- RIGHT JOIN --------
right_df = pd.merge(students_df, enrollments_df,
                    on='student_id', how='right')

print("Right Join Result ({} rows):".format(len(right_df)))
print(right_df[['student_id', 'name', 'course_name', 'enrollment_date']])

no_names = right_df[right_df['name'].isnull()]['student_id'].tolist()
print(f"Student_ids without names: {no_names}\n")


# -------- OUTER JOIN --------
outer_df = pd.merge(students_df, enrollments_df,
                    on='student_id', how='outer')

print("Outer Join Result ({} rows):".format(len(outer_df)))
print(outer_df[['student_id', 'name', 'course_name', 'enrollment_date']])

problem_rows = outer_df[
    outer_df['name'].isnull() | outer_df['course_name'].isnull()
]

print("\nRows with missing name OR course_name:")
print(problem_rows[['student_id', 'name', 'course_name']])
print()


# -------- OUTER JOIN WITH INDICATOR --------
outer_indicator = pd.merge(
    students_df, enrollments_df,
    on='student_id',
    how='outer',
    indicator=True
)

print("Merge Source Distribution:")
print(outer_indicator['_merge'].value_counts())
print()


# ===============================
# === TASK 3: Lookup & Automation ===
# ===============================

print("=== TASK 3: Lookup and Automation ===\n")

# Create Scores DataFrame
scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

scores_df = pd.DataFrame(scores_data)

# Lookup using map
score_dict = dict(zip(scores_df['student_id'], scores_df['exam_score']))
students_df['exam_score'] = students_df['student_id'].map(score_dict)

print("Lookup Operation Result:")
print(students_df[['student_id', 'name', 'exam_score']])
print()


# ===============================
# Automation Function
# ===============================

def auto_merge(df1, df2, join_type, key_column):
    merged_df = pd.merge(df1, df2,
                         on=key_column,
                         how=join_type)

    return {
        'result_df': merged_df,
        'row_count': len(merged_df),
        'join_type': join_type
    }

# Test 1 - Inner Join
result_inner = auto_merge(students_df, enrollments_df, 'inner', 'student_id')

print("Automation Function Test:")
print("Join Type:", result_inner['join_type'])
print("Rows in Result:", result_inner['row_count'])
print("Result Preview:")
print(result_inner['result_df'][['student_id','name','course_name']].head())
print()

# Test 2 - Left Join
result_left = auto_merge(students_df, enrollments_df, 'left', 'student_id')

print("Join Type:", result_left['join_type'])
print("Rows in Result:", result_left['row_count'])
print("Result Preview:")
print(result_left['result_df'][['student_id','name','course_name']].head())
```
